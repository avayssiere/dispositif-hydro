```python
import os
import pandas as pd

data_folder = "data"
os.listdir(data_folder)
```

```python
q_m3s_SAM = pd.read_excel(os.path.join(data_folder, "StA_d\x82bit_2015-2018.xlsx"), skiprows=1, index_col=2)
q_m3s_SAM.set_index(pd.to_datetime(q_m3s_SAM.index), inplace=True)
# q_m3s_SAM = q_m3s_SAM.loc[:, "Debit en m3/s"]
q_m3s_SAM.drop("NaT", inplace=True)

q_m3s_SAM.head()
```

```python
pre_martial_df = pd.read_excel(os.path.join(data_folder, "Pre_Martial_modele.xlsx"), skiprows=8, index_col=0)
pre_martial_df.set_index(pd.to_datetime(pre_martial_df.index), inplace=True)
pre_martial_df = pre_martial_df.iloc[:, [0, 1, 2, 5]]
pre_martial_df.head()
```

```python
arnet_df = pd.read_excel(os.path.join(data_folder, "Arnet_modele.xlsx"), skiprows=8, index_col=0)
arnet_df.set_index(pd.to_datetime(arnet_df.index), inplace=True)
arnet_df = arnet_df.iloc[:, [0, 1, 2, 5]]
arnet_df.rename(columns={"Alti NGF toit de la nappe": "Alti NGF"}, inplace=True)
arnet_df.head()
```

```python
le_cher_df = pd.read_excel(os.path.join(data_folder, "Le_cher_modele.xlsx"), skiprows=8, index_col=0)
le_cher_df.set_index(pd.to_datetime(le_cher_df.index), inplace=True)
le_cher_df = le_cher_df.iloc[:, [0, 1, 2, 5]]
le_cher_df.rename(columns={le_cher_df.columns[-1]: "Alti NGF"}, inplace=True)
le_cher_df.head()
```

```python
lp1_df = pd.read_excel(os.path.join(data_folder, "LP1_modele.xlsx"), skiprows=8, index_col=0)
lp1_df.set_index(pd.to_datetime(lp1_df.index), inplace=True)
lp1_df = lp1_df.iloc[:, [0, 1, 2, 5]]
lp1_df.rename(columns={lp1_df.columns[-1]: "Alti NGF"}, inplace=True)
lp1_df.head()
```

```python
chevalier_df = pd.read_excel(os.path.join(data_folder, "Chevalier_modele.xlsx"), skiprows=8, index_col=0)
chevalier_df.set_index(pd.to_datetime(chevalier_df.index), inplace=True)
chevalier_df = chevalier_df.iloc[:, [0, 1, 2, 5, 6]]
chevalier_df.rename(columns={"Alti NGF toit de la nappe": "Alti NGF"}, inplace=True)
chevalier_df.head()
```

```python
meteo_orval = pd.read_excel(os.path.join(data_folder, "m\x82t\x82o_orval.xlsx"), index_col=1)
meteo_orval.set_index(pd.to_datetime(meteo_orval.index), inplace=True)
meteo_orval = meteo_orval.iloc[:, [1, 3]]
meteo_orval.rename(columns={"RR-précipitation (mm)": "RR(mm)","TM -Température (C°)": "TM(C°)"}, inplace=True)
meteo_orval.head()
```

```python
from sklearn import preprocessing
```

```python
end = 30
results = {}
temp_to_shift = temp.copy().resample("1D").mean()
temp_to_shift.dropna(inplace=True)

for col in temp_to_shift.columns:
    min_max_scaler = preprocessing.MinMaxScaler()
    temp_to_shift[col] = min_max_scaler.fit_transform(temp_to_shift[col].values.reshape(-1, 1))

for i in range(1, end):
    temp_to_shift = temp_to_shift.shift(1)
    results[i] = pd.concat([temp_to_shift, meteo_orval["TM(C°)"]], axis=1).corr()["TM(C°)"]

results[10]
```

```python
res_corr = {k: v.drop("TM(C°)").to_dict() for k, v in results.items()}
res_corr_df = pd.DataFrame().from_dict(res_corr, orient="index")
res_corr_df.head()
```

```python
plt.figure(figsize=(9, 6))

for col in res_corr_df.columns:
    plt.plot(res_corr_df[col], '-', label=col, alpha=0.75)
```
